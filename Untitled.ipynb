## Singly Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None 
        
class LinkedList:
    def __init__(self):
        self.head = None
```

```python
## INSERTION

# Insertion at begining
def push(self, new_data):
    new_node = Node(new_data)
    new_node.next = self.head
    self.head = new_node
    
# Insertion after the given prev_node.
def insertAfter(self, prev_node, new_data):
    if prev_node is None:
        print("The given previous node must inLinkedList.")
        return
    new_node = Node(new_data)
    new_node.next = prev_node.next
    prev_node.next = new_node
    
# Insertion at end.
def append(self, new_data):
    new_node = Node(new_data)
    if self.head is None:
        self.head = new_node
        return
    last = self.head
    while (last.next):
        last = last.next
    last.next =  new_node
```

```python
## DELETION
# Delete a particular value
def deleteKey(self, key):
    temp = self.head
    if (temp is not None):
        if (temp.data == key):
            self.head = temp.next
            temp = None
            return
    while(temp is not None):
        if temp.data == key:
            break
        prev = temp
        temp = temp.next
    if(temp == None):
        return
    prev.next = temp.next
    temp = None
# Delete node at particular position
def deletePos(self, position):
    if self.head == None:
        return
    temp = self.head
    if position == 0:
        self.head = temp.next
        temp = None
        return
    for i in range(position -1 ):
        temp = temp.next
        if temp is None:
            break
    if temp is None:
        return
    if temp.next is None:
        return
    next = temp.next.next
    temp.next = None
    temp.next = next
# Delete complete linkedlist
def deleteList(self):
    current = self.head
    while current:
        prev = current.next
        del current.data
        current = prev
```

```python
## LENGTH
# Iterative method
def getCount(self):
        temp = self.head
        count = 0
        while (temp):
            count += 1
            temp = temp.next
        return count
# Recursive method
def getCountRec(self, node):
        if (not node):
            return 0
        else:
            return 1 + self.getCountRec(node.next)
```

```python
## SEARCH IF THE NODE IS PRESENT OR NOT
# Iterative method
def search(self, x):
    current = self.head
    while current != None:
        if current.data == x:
            return True            
        current = current.next
    return False
# Recursive method
def searchRec(self, li, key):
        if(not li):
            return False
        if(li.data == key):
            return True
        return self.search(li.next, key)
```
